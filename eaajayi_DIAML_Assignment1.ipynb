```python
# Question 5

import pandas as pd
import numpy as np

capital = 100000
customer_base = 100
interest = 0
day = 0
days = [day] # Initialize list of days with value 0
total_interest = [interest] # Initialize list of accumulated interest with value 0


while total_interest[-1] < 100000:
  day+=1
  days.append(day)
  customer_base*=1.01
  interest+= customer_base * 10
  total_interest.append(interest)


# print(days)
# print(total_interest)

interest_data = pd.DataFrame({
    "Day": days,
    "Total Interest": total_interest
})

interest_data




```

```python
import matplotlib.pyplot as plt

# Plot the DataFrame
plt.figure(figsize=(10, 6))
plt.plot(interest_data['Day'], interest_data['Total Interest'], linestyle='-', color='b')

plt.axvline(x=interest_data['Day'].max(), color='r', linestyle='--', linewidth=1, label='End of Period')

# Add titles and labels
plt.title('Amount Over Time')
plt.xlabel('Day')
plt.ylabel('Total Interest ($)')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
